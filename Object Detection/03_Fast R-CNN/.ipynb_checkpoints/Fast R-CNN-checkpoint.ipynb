### Backgroud

- R-CNN은 2014년 major breakthrough 였다.  


- 하지만, R-CNN 에는 몇가지 문제가 있었다.
    - SLOW: 각각의 regional proporsal(2000개 정도)에 대해서 계산을 해줘야했다.
    
    - Hard to train: 훈련시 세 단계를 거쳐야하는 어려움이있었다. (CNN, SVM, BB regression)
    
    - Large Memory: 각각의 regional proposal에 대해서 모든 피쳐맵을 저장해야만 하기 때문에 메모리가 매우 많이 든다.
    

=> Fast R-CNN이 나온 계기가 되었다.

### Fast R-CNN

- 위의 문제점들을 해결하기위해 저자가 새로운 방법을 제시.
    - Fast R-CNN의 주요점 -> We only have one system, that we can train end-to-end
    - 즉, detection하기 위한 세 가지 작업이였던 (CNN, SVM, BB regressor)를 하나의 구조로 합친 것!

### Structure

- 1. 전체 이미지를 CNN으로 처리한다. output은 `하나의 featuremap`이다. *  이 부분 다시 확인


- 2. 각각의 region proposal은 그 feature map에서 해당하는 부분을 추출한다.
    - 이것을 region proposal feature map이라고 한다.
    - 자세히는, feature map에서 region proposal에 해당하는 부분을 가져온다음, pooling layer를 통해서 고정된 사이즈로 만든다.
        - 이때, pooling layer는 Region of Interest로 (ROI) pooling layer라고 한다.
        -> https://deepsense.ai/region-of-interest-pooling-explained/

- 3. 그 후, fixed size의 region proposal feature map을 faltten 시킨다.
    - 이 flatten된 벡터가 feature vector가 되므로 항상 같은 사이즈를 유지한다.
    
    
- 4. 이 feature vector가 두 개의 FC layer의 input으로 사용된다.
    - 첫번 째는 softmax classification을 위한 FC layer이다.
    - 두번 째는 Bounding Box regressor이며, 물체의 coordinates를 예측한다.

### Conclusion

- 두 논문이 나옴으로써, object detection계에 엄청난 향상을 가져왔다.


- 특히 inference time(추론 시간)의 향상이 눈에 띄었으며, 결국 real-time에 사용된다.
