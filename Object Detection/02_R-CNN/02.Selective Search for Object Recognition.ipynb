# What is Selective Search?

- Region Proposals 알고리즘 중 하나

- 색상, 질감, 크기 와 compatible한 모양에 근거해 비슷한 지역을 계층적으로 그룹화하여 계산하는 알고리즘.


- Selective Search (SS)는 `Felzenszwalb and Huttenlocher`의 segmentation 방법인 그래프방법을 이용해 픽셀의 강도에 근거해 이미지를 over-segmenting하는 방법이다.
    -  segmentation방법의 결과

- 위에 보이는 output image를 region proposals로 그대로 사용해도 될까?  
    - No
        - 대부분의 실제 물체들이 2개 또는 그 이상의 segmented된 부분을 갖고있기 때문.
        - cup에 의해 가려진 plate같은 물체들에 대한 Region Proposal들은 해당 방법으로는 생성되지 않는다.
    

- 첫 번째 문제를 해결학 위해, 지역에 가장가까운 비슷한 지역들을 병합하다보면 두개의 물체를 하나의 지역으로 segmentation 할수 도 있는 문제가 있다.
    - 현재 해결하려는 문제느 Perfect segmentation이 아니기 때문에 최대한 많은 region proposal들을 예측하려는 것이다.
    
    
- Felzenszwalb and Huttenlocher의 segmentation방법을 이용한 `oversegment`한 결과


### Selective Search 알고리즘의 절차

- input: oversegment된 이미지


1.Regional proposal의 리스트에 따라 segmented된 부분들에 맞게 모든 bounding boxes를 추가

2.similarity 에 근거한 가장 가까운 segment들 그룹화

3.stpe 1 으로 다시 돌아가서 Iteration 반복


- 매 iteration 마다, 더욱 큰 segment들이 형성되고, region proposal에 추가된다.


- 즉, 작은 segments들로부터 큰 segments들로 region proposal들이 만들어지며 bottom-up approach라고 보면 된다.


- 이 때문에 `hierarchial` segmentation이라고 한다.

#### Similarity

- 두개의 region에 대해서 similarity를 계산하는 방법은 4가지가 있다.
    - color; 두 지역간의 color histogram 이용 (bins은 25)


- texture; 각 채널의 8 orientatino에서 가우시안 미분을 한 것을 비교


- size


- shape compatibility


#### Final Similarity

- SS에서 similarity를 계산하는 방법은 위의 네 가지 알고리즘 중 하나의 선형조합으로 정의하여 사용하며 두 지역간의 similarity를 계산한다.



- $r_i$와 $r_j$는 이미지내의 두 지역이나 segments를 의미한다

- a1 ~a4는 해당 방법이 사용되었는지 아닌지를 의미한다. 
    - 사용되면 1 아니면 0

#### 결과

- SS 알고리즘에 의해 생성된 boxes들 중, top 200-250을 사용한 결과이다.



- 보통 1000-1200개의 proposals이면 correct region proposal을 얻을 수 있다.
